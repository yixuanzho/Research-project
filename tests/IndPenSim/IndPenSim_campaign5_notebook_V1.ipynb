The IndPenSim_V2_export_V7_CampaignALL_Campaign5.csv and IndPenSim_V2_export_V7_CampaignALL_Campaign5_Statistics.csv datasets used in this Jupyter notebook were generated by the MATLAB script Generate_Production_Batch_data_V4_CampaignALL.m. 

This script is based on the original file Generate_Production_Batch_data_V4.m, with the addition of a loop statement designed to simulate data for a five-year production period as described in the paper. The data is divided into five annual campaigns, each employing different control strategies and either fixed or variable batch lengths. The authors did not consider any advanced process control strategies, such as Raman spectroscopy. For details, please refer to Table 2 in the paper.

This Jupyter notebook (or MATLAB script) aims to generate summary statistics and reproduce Figure 4 from the paper. The data processing has already been completed.

```python
if all(campaign5.Batch_ref==0):
    campaign5 = campaign5.drop('Batch_ref',axis=1)
    campaign5 = campaign5.rename(columns={'Batch_ref.1': 'Batch_ref'})
```

```python
variable_list_old = campaign5.columns
variable_list_old
```

```python
variable_list_new = [
    'Time (h)', 'Aeration rate(Fg:L/h)', 'Agitator RPM(RPM:RPM)',
    'Sugar feed rate(Fs:L/h)', 'Acid flow rate(Fa:L/h)',
    'Base flow rate(Fb:L/h)', 'Heating/cooling water flow rate(Fc:L/h)',
    'Heating water flow rate(Fh:L/h)',
    'Water for injection/dilution(Fw:L/h)',
    'Air head pressure(pressure:bar)', 'Dumped broth flow(Fremoved:L/h)',
    'Substrate concentration(S:g/L)',
    'Dissolved oxygen concentration(DO2:mg/L)',
    'Penicillin concentration(P:g/L)', 'Vessel Volume(V:L)',
    'Vessel Weight(Wt:Kg)', 'pH(pH:pH)', 'Temperature(T:K)',
    'Generated heat(Q:kJ)',
    'carbon dioxide percent in off-gas(CO2outgas:%)',
    'PAA flow(Fpaa:PAA flow (L/h))',
    'PAA concentration offline(PAA_offline:PAA (g L^{-1}))',
    'Oil flow(Foil:L/hr)',
    'NH_3 concentration off-line(NH3_offline:NH3 (g L^{-1}))',
    'Oxygen Uptake Rate(OUR:(g min^{-1}))',
    'Oxygen in percent in off-gas(O2:O2  (%))',
    'Offline Penicillin concentration(P_offline:P(g L^{-1}))',
    'Offline Biomass concentratio(X_offline:X(g L^{-1}))',
    'Carbon evolution rate(CER:g/h)', 'Ammonia shots(NH3_shots:kgs)',
    'Viscosity(Viscosity_offline:centPoise)',
    'Fault reference(Fault_ref:Fault ref)',
    '0 - Recipe driven 1 - Operator controlled(Control_ref:Control ref)',
    '2-PAT control(PAT_ref:PAT ref)',
    'Batch reference(Batch_ref:Batch ref)',
]
```

```python
# Create a mapping dictionary from old column names to new column names
column_mapping = {}
for i in range(len(variable_list_old)):
    column_mapping[variable_list_old[i]] = variable_list_new[i]
# Rename the columns using the dictionary
campaign5 = campaign5.rename(columns=column_mapping)
```
